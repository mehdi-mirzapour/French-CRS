```python
import json
import pandas as pd
import glob
import re
import random
import glob
import random
```

```python
ANCOR_dataframe=pd.read_excel("./corpus_OTG_normal_Neg_50.xlsx",index_col=0)
ANCOR_dataframe
```

```python
sub_corpus_num=0
rate=0

while (sub_corpus_num!=2 or not(rate>=.2000 and rate<.22) ):
    sub_corpus_files=ANCOR_dataframe["File_Name"].unique().tolist()
    sub_corpus_files_test=random.sample(sub_corpus_files, 45)
    sub_corpus_files_train=[file for file in sub_corpus_files if file not in sub_corpus_files_test]


    ANCOR_dataframe_train=ANCOR_dataframe[ANCOR_dataframe["File_Name"].isin(sub_corpus_files_train)]
    ANCOR_dataframe_test=ANCOR_dataframe[ANCOR_dataframe["File_Name"].isin(sub_corpus_files_test)]

    rate=ANCOR_dataframe_test.shape[0]/ANCOR_dataframe_train.shape[0]
    sub_corpus_num=(ANCOR_dataframe_train["Sub_Corpus"].unique().shape[0]+
                    ANCOR_dataframe_test["Sub_Corpus"].unique().shape[0])
    
print(len(sub_corpus_files_test))
print(len(sub_corpus_files_train))
print(rate)
print(sub_corpus_num)
```

```python
ANCOR_dataframe_train.to_excel("./All-Chains-Balanced-Links/corpus_ALL_train.xlsx")
ANCOR_dataframe_test.to_excel("./All-Chains-Balanced-Links/corpus_ALL_test.xlsx")
```
