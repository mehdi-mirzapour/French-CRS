# Creating General ANCOR Dataset

```python
from ancor2dataset import chains2excel
import glob
import pandas as pd

PATH_ANCOR="../../DISTRIB_ANCOR/"
SUB_CORPRA=["corpus_OTG","corpus_UBS","corpus_ESLO","corpus_ESLO_CO2"]

for SUB_CORPUS in SUB_CORPRA:

    PATH_ANCOR_SUB_CORPUS=PATH_ANCOR+SUB_CORPUS+'/'

#     all_files=[file[file.rfind("/")+1:-4] for file in glob.glob(PATH_ANCOR_SUB_CORPUS+"annotation_integree/*.xml")]
#     filter_files=dict_files["sub_corpus_files_test"]
    
#     corpus_files=[value for value in all_files if value in filter_files] 

    corpus_files=[file[file.rfind("/")+1:-4] for file in glob.glob(PATH_ANCOR_SUB_CORPUS+"annotation_integree/*.xml")]

    
    list_dataframes=[]


    for file in corpus_files:

        dataset = chains2excel(PATH_ANCOR)
        print("Starting",file)
        dataset.set_file_config(SUB_CORPUS, file)
        dataset.generate_json_mentions()
        dataset.remove_empty_json_mentions()
        dataset.generate_json_relations()
        if dataset.json_relations=={}:
            continue
        dataset.generate_json_chains()
#         dataset.generate_pos_neg_pairs(neg_pairs_percentage=0.5,
#                                  strategy="normal")

        dataset.generate_pos_neg_pairs(strategy = "pairs_from_left_with_window_size", window_size=5)
    
        dataset.extend_with_order_pos_neg_pairs()
        dataset.generate_training_test_dataset(save_files=False)


        list_dataframes.append(dataset.coreference_pairs_dataframe)


    train_test_dataframe=pd.concat(list_dataframes,ignore_index=True)
    train_test_dataframe.to_excel("./ds_window_15/"+SUB_CORPUS+"_win.xlsx")
```

```python
PATH="./ds_window_15/"
merge_lists=[]
files_list=[file[file.rfind("/")+1:] for file in glob.glob(PATH+"*.xlsx")]
# files_list=["corpus_UBS.xlsx","corpus_OTG.xlsx"]

for file in files_list:
    merge_lists.append(pd.read_excel(PATH+file,index_col=0))

ANCOR_dataframe=pd.concat(merge_lists,ignore_index=True)
ANCOR_dataframe.to_excel(PATH+"corpus_ANCOR_Window_5.xlsx")    
```
